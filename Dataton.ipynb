# 2) Projeto: API Open Weather Map

Este projeto de API Open Weather Map tem como objetivo retirar informações de tempo atuais de cidades Brasileiras!

Importando as bibliotecas necessárias para tratamento das informações

```python
import json
import requests
import pandas as pd
import numpy as np
import xlsxwriter
```

Obtivemos um arquivo json com o nome de todas as cidades.
O arquivo json city_list.json foi obtido no próprio site http://bulk.openweathermap.org/sample/city.list.json.gz

Como este arquivo está zipado, baixamos ele diretamente no mesmo diretório deste jupyter notebook e tratamos ele a partir daí.

```python
# Lendo arquivo de cidades do Open Weather
json_file = pd.read_json('city_list.json', encoding='utf-8')

# Salvando todos os registros de cidades brasileiras
json_file = json_file[json_file['country']=='BR']

# Salvando somente os nomes de cidades brasileiras
cities_list = list(json_file['name'])
cities_list
```

Tivemos que selecionar apenas 50 cidades do Brasil, pois o Jupyter não aguenta processar muitas cidades em uma iteração :(

Um ponto importante é que as cidades não estão divididas por estado, portanto, há nomes de cidades repetidos.

Este código faz a chamada via resquests.get() para o website do Open Weather Map, passando como parâmetro as cidades brasileiras selecionadas.

Como resultado final, teremos um dicionario para cada cidade, com informações de tempo. Tudo salvo na lista 'br_cities'.

```python
br_cities = []
lista=cities_list[0:50]
for city in lista:
    try:
        url = 'http://api.openweathermap.org/data/2.5/weather?q={}&appid=3064e81432c51c7940eaa8dbe01ff65f'.format(city)
        response = requests.get(url)
        data = response.json()
        br_cities.append( data)
        print('city: {city} response code: {status}'.format(city=city, status=response))
    except:
        print('Can not retrieve weather information for the city: ', city)
```

Imprimindo as informações de tempo das cidades selecionadas

```python
br_cities
```

Definindo uma função 'redefine_dict()' que irá remodelar toda a estrutura de dados de tempo para cada cidade. Ao invés de conter tipos diferentes de valores (lista, dicionario e string), os valores serão somente do tipo string. Neste caso, as keys dos dicionarios encadeados farão parte da key do dicionario mãe. 

Conforme exemplo:
A estrutura do dicionario para a cidade de São Paulo é:

Após a transformação da função abaixo, ela ficará:

Definindo a função redefine_dict(), que recebe como argumento um dicionario cujos valores têm que ser do tipo list, dict ou string. 

Retorna um dicionario com valores do tipo string somente.

As subkeys não são perdidas, mas incorporadas à key 'mãe'.

```python
def redefine_dict(dict_city):
    dict_temp = {}
    for key,value in dict_city.items():
        if isinstance(value, list):
            for sub_value in value:
                for sub_v_key, sub_v_value in sub_value.items():
                    dict_temp['_'.join([str(key), str(sub_v_key)])] = sub_v_value
        elif isinstance(value, dict):
            for sub_v_key, sub_v_value in value.items():
                dict_temp['_'.join([str(key), str(sub_v_key)])] = sub_v_value
        elif isinstance(value, str):
            dict_temp[key] = value
    return dict_temp
```

Aplicando a remodelação da estrutura do dicionario contendo as informações de tempo das cidades selecionadas, através da chamada à função 'redefine_dict'.

```python
redefined_cities = []

for city in br_cities:
    redefined_cities.append(redefine_dict(city))

#imprimindo o primeiro elemento da lista, somente como amostra
print(redefined_cities[0])
```

Salvando as listas de cidade e previsão do tempo atual em um DataFrame com o nome 'df_weather_brcities'

```python
colunas = list(redefined_cities[0].keys())
df_weather_brcities = pd.DataFrame(redefined_cities, columns=colunas)
df_weather_brcities.head(15)
```

Salvando o DataFrame em um arquivo .csv

```python
# salvando DataFrame em formato csv
df_weather_brcities.to_csv('BR_Cities_Weather.csv', encoding='utf-8')
```

Definindo a função 'redefine_dict_2', que é quase igual a função anteriormente apresentada (redefine_dict).

O que muda: caso a key do dicionário seja igual a 'qsa', ela adiciona o valor inteiro (sem particionar o dict) para o dict_temp() que vai ser retornado no final da função. Isso foi feito pois esta informação não tem um padrão fácil de encontrar para redefini-la em novas colunas.

```python
def redefine_dict_2(dict_city):
    dict_temp = {}
    for key,value in dict_city.items():
        if key == 'qsa':
            dict_temp[key] = value
        elif isinstance(value, list):
            for sub_value in value:
                for sub_v_key, sub_v_value in sub_value.items():
                    dict_temp['_'.join([str(key), str(sub_v_key)])] = sub_v_value
        elif isinstance(value, dict):
            for sub_v_key, sub_v_value in value.items():
                dict_temp['_'.join([str(key), str(sub_v_key)])] = sub_v_value
        elif isinstance(value, str):
            dict_temp[key] = value
    return dict_temp
```

# 4) Projeto: Webscraping Adoro Cinema (Estréias)

O objetivo desse projeto é obter a lista de filmes em cartaz, disponível no site Adoro Cinema

Vamos fazer uma requisição para o site adorocinema.com e salvar as informações na variável "page".
Depois disso, vamos utilizar o BeautiulSoup para obter nossos dados mais bonitinhos.

```python
# obtendo os dados do site
page = requests.get('http://www.adorocinema.com/filmes/numero-cinemas')
soup = BeautifulSoup(page.content, 'html.parser')
```

Agora vamos extrair as informações de títulos dos filmes, que estão na tag "a"  

```python
# extraindo os títulos de estréia

titles = [i.text
       for i in soup.find_all('a', {'class': 'meta-title-link'})]
titles
```

Agora vamos obter as informações de data da estréia para cada filme

```python
# extraindo as datas de estréia de cada título

dates = [i.text
       for i in soup.find_all('span', {'class': 'date'})]
dates
```

Em seguida, extrairemos os diretores para cada filme

```python
# extraindo os diretores de cada título

director = [i.text
       for i in soup.find_all('a', {'class': 'blue-link'})]
director
```

O próximo passo será obter a descrição de cada filme

```python
# extraindo a descrição de cada título

desc = [i.text.strip('\n').strip().strip('\n')
       for i in soup.find_all('div', {'class': 'content-txt'})]
```

Agora vamos unir todas as listas em um DataFrame

```python
# unindo as listas de título e data em um data frame

df1 = pd.DataFrame(list(zip(titles, dates)),
              columns =['Títulos', 'Data de estréia'])

# unindo as listas de diretor e descrição em um data frame

df2 = pd.DataFrame(list(zip(director, desc)),
              columns =['Diretor', 'Descrição'])

# concatenando os data frames criados anteriormente

df_adoro_cinema = pd.concat([df1, df2], axis=1)
df_adoro_cinema
```

Finalmente, vamos salvar as informações do DataFrame no arquivo "Filmes_Em_Cartaz.csv"

```python
df_adoro_cinema.to_csv('Filmes_Em_Cartaz.csv', encoding='utf-8')
```

# 5) Projeto: Webscraping Top 100 Músicas da Billboard

O objetivo deste projeto é obter as TOP 100 músicas da Billboard.

Primeiro vamos fazer o request para o site da Billboard e salvar esta informação utilizando o BeautifulSoup para deixar mais legível os dados e podermos percorrer as tags.

```python
# obtendo os dados do site

page = requests.get('https://www.billboard.com/charts/hot-100')
soup = BeautifulSoup(page.content, 'html.parser')
```

Agora vamos obter o nome das músicas mais ouvidas

```python
# extraindo as músicas mais ouvidas

musicas = [i.text
       for i in soup.find_all('span', {'class': 'chart-element__information__song text--truncate color--primary'})]
```

Extraindo o nome dos artistas para cada música

```python
# extraindo os artistas correspondentes às músicas

artistas = [i.text
       for i in soup.find_all('span', {'class': 'chart-element__information__artist text--truncate color--secondary'})]
```

Obtendo o ranking para cada música

```python
# extraindo posição mais alta ocupada pela música

peak = [i.text
       for i in soup.find_all('span', {'class': 'chart-element__meta text--center color--secondary text--peak'})]
```

Obtendo quantas semanas a música ficou no top 100

```python
# extraindo quantas semanas a música ficou no top 100

duration = [i.text
       for i in soup.find_all('span', {'class': 'chart-element__meta text--center color--secondary text--week'})]
```

Salvando tudo em um DataFrame

```python
# unindo as listas de música e artistas em um data frame

df1 = pd.DataFrame(list(zip(musicas, artistas)),
              columns =['Música', 'Artista'])

# unindo as listas de pico e duração em um data frame

df2 = pd.DataFrame(list(zip(peak, duration)),
              columns =['Posição mais alta', "Duração (em semanas)"])

# concatenando os data frames criados anteriormente

df3 = pd.concat([df1, df2], axis=1)
df3.index = df3.index + 1
df3
```

Finalmente, vamos salvar no arquivo Top100Billboard.csv

```python
df3.to_csv('Top100Billboard.csv', encoding='utf-8')
```
